```python
import pandas as pd
import numpy as np
import math as mat
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('p1_dataset_elpino_raw.csv', sep=';', na_values='-')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Sexo (Desc)'] = df['Sexo (Desc)'].replace('Hombre', "1")
df['Sexo (Desc)'] = df['Sexo (Desc)'].replace('Mujer', "0")
df['Sexo (Desc)']=pd.to_numeric(df['Sexo (Desc)'], errors='coerce')

numerical_features = ["Edad en años", "Sexo (Desc)"]

categorical_features = df.columns.values.tolist()

categorical_features.remove("Edad en años") 
categorical_features.remove("Sexo (Desc)")
```

<h1> Outliers sin eliminar duplicados

```python
df.duplicated().sum()
```

```python
#Eliminamos esos datos repetidos
df = df.drop_duplicates()
```

<h1> Tabla de frecuencias categoricas

```python
# Itera a través de cada variable categórica y muestra su tabla de frecuencia
for col in categorical_features:
    print(f"Tabla de frecuencia para la variable {col}:")
    print(pd.DataFrame(df[col].value_counts()))
    print("\n")
```

<h1> Tabla de frecuencias numericas

```python
# Itera a través de cada variable categórica y muestra su tabla de frecuencia
for col in numerical_features:
    print(f"Tabla de frecuencia para la variable {col}:")
    print(pd.DataFrame(df[col].value_counts()))
    print("\n")
```

```python
print("Promedio de edad \n")
print(df['Edad en años'].mean())
```

```python
edades=df['Edad en años']
green_diamond = dict(markerfacecolor='g', marker='D')
fig, ax = plt.subplots()
ax.set_title('Boxplot por Edades')
ax.boxplot(edades, flierprops=green_diamond, labels=["Edad"])
```
