```python
import pandas as pd
import numpy as np
import math as mat
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('p1_dataset_elpino_raw.csv', sep=';', na_values='-')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Sexo (Desc)'] = df['Sexo (Desc)'].replace('Hombre', "1")
df['Sexo (Desc)'] = df['Sexo (Desc)'].replace('Mujer', "0")
df['Sexo (Desc)']=pd.to_numeric(df['Sexo (Desc)'], errors='coerce')

numerical_features = ["Edad en años", "Sexo (Desc)"]

categorical_features = df.columns.values.tolist()

categorical_features.remove("Edad en años") 
categorical_features.remove("Sexo (Desc)")
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
mean = df['Edad en años'].mean()
std = df['Edad en años'].std()

limit = mean + 3*std
df.loc[df['Edad en años'] > limit, 'Edad en años'] = np.nan

df['Edad en años'].fillna(mean, inplace=True)
```
