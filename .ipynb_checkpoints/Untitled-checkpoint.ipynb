```python
import pandas as pd
import numpy as np
import math as mat
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('p1_dataset_elpino_raw.csv', sep=';', na_values='-')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Sexo (Desc)'] = df['Sexo (Desc)'].replace('Hombre', "1")
df['Sexo (Desc)'] = df['Sexo (Desc)'].replace('Mujer', "0")
df['Sexo (Desc)']=pd.to_numeric(df['Sexo (Desc)'], errors='coerce')

numerical_features = ["Edad en años", "Sexo (Desc)"]

categorical_features = df.columns.values.tolist()

categorical_features.remove("Edad en años") 
categorical_features.remove("Sexo (Desc)")
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
mean = df['Edad en años'].mean()
std = df['Edad en años'].std()

limit = mean + 3*std
df.loc[df['Edad en años'] > limit, 'Edad en años'] = np.nan

df['Edad en años'].fillna(mean, inplace=True)
```

```python
# Itera a través de cada variable categórica y muestra su tabla de frecuencia
for col in categorical_features:
    print(f"Tabla de frecuencia para la variable {col}:")
    print(pd.DataFrame(df[col].value_counts()))
    print("\n")
```

```python
# Crea una figura con subplots para cada variable categórica
fig, axes = plt.subplots(nrows=len(categorical_features), figsize=(8, 30))

# Itera a través de cada variable categórica y grafica su tabla de frecuencia
for i, col in enumerate(categorical_features):
    # Crea una tabla de frecuencia para la variable actual
    freq_table = df[col].value_counts()
    
    # Grafica la tabla de frecuencia en el subplot correspondiente
    ax = freq_table.plot(kind='bar', ax=axes[i], rot=0, color='blue')
    ax.set_title(f"Tabla de frecuencia para la variable {col}")
    ax.set_xlabel(col)
    ax.set_ylabel("Frecuencia")
    ax.grid(axis='y')

# Ajusta los subplots para que no se superpongan
plt.tight_layout()

# Muestra la figura
plt.show()
```
